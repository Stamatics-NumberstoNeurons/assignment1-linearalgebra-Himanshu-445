### Q3 Part (i) — Similarity Matrix

What linear algebra concept makes this possible in one line?

````python
# Input — do not modify
users = np.array([
    [0.9, 0.1, 0.8, 0.2],  # User A
    [0.1, 0.9, 0.2, 0.8],  # User B
    [0.8, 0.2, 0.9, 0.1],  # User C
    [0.5, 0.5, 0.5, 0.5],  # User D
])
````

### Q3 Part (iii) — The Problem with Raw Dot Products

A senior engineer tells you:
> *"Raw dot products are a bad measure of similarity — you need to normalize."\*

1. What is the **problem** with raw dot products?
2. Construct a **specific example using NumPy** that demonstrates the flaw.
3. Fix it.

> **Hint:** Take a user with scores `[1.8, 0.2, 1.6, 0.4]`

*Write your answer here (double-click to edit this cell)*

````python
# Write your code here:
user_a = np.array([0.9, 0.1, 0.8, 0.2])
user_new = np.array([1.8, 0.2, 1.6, 0.4])

print("Raw dot product (A and A):", np.dot(user_a, user_a))
print("Raw dot product (A and New User):", np.dot(user_a, user_new))

def cosine_sim(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

print("Normalized similarity (A and A):", cosine_sim(user_a, user_a))
print("Normalized similarity (A and New User):", cosine_sim(user_a, user_new))
````

### Q3 Part (iv) — Most Similar Users

After normalizing, which **two users are most similar**?

Does that match your intuition from just reading the vectors?

What does it mean in plain language — what do these two users have in common?

> **Note:** A similarity score is just a number that tells you **how alike two things are**.

*Write your answer here (double-click to edit this cell)*

````python
# Write your code here:
norms = np.linalg.norm(users, axis=1, keepdims=True)
normalized_users = users / norms

cosine_sim_matrix = normalized_users @ normalized_users.T

print("Normalized Cosine Similarity Matrix:\n")
np.set_printoptions(precision=3, suppress=True)
print(cosine_sim_matrix)
````
